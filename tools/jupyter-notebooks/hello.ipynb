```python
# Display a dot graph.
from graphviz import Source

dotGraphString = """
digraph G {
    a -> b;
}
"""
Source(dotGraphString)
```

```python
# Create and save the dot graph to a file.
dotGraphString = """
digraph G {
    a -> b;
}
"""
# Write the dot graph to a file.
dotGraphFile = "hello.dot"
with open(dotGraphFile, "w") as f:
    f.write(dotGraphString)
```

```python
# Display the dot graph from a file.
Source.from_file(dotGraphFile)
```
